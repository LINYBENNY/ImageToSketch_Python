#  Importing Required libraries

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

# Loading the Image

```python
image=cv2.imread("Lion.jpg")
image
```

```python
image = cv2.resize(image,(500,300))
image
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Converting into Gray Scale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

# Inverting the image

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
```

```python
plt.imshow(inverted_image)
```

# Smoothing the image

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
img_smoothing
```

```python
plt.imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

# Inverting the Image to get the Pencil Sketch

```python
pencil_sketch = dodgeV2(gray_image, img_smoothing)
plt.imshow(pencil_sketch)
```

# Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```
